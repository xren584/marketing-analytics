```python
#delete the rows with null value in join_fee column
subscriber.drop('Unnamed: 0', axis=1, inplace=True)
subscriber.dropna(subset=['join_fee'],inplace=True)
subscriber.info()
```

```python
subscriber['join_fee'].value_counts()
subscriber.drop(subscriber[subscriber.join_fee < 0].index, inplace=True)
```

```python
subscriber.info()
subscriber['join_fee'].value_counts()
subscriber['trial_completed'].value_counts()
```

```python
Group_no_trial_fee=subscriber[(subscriber['join_fee']==0) & (subscriber['trial_completed']== True)]
Group_low_trial_fee=subscriber[(subscriber['join_fee']==0.0367) & (subscriber['trial_completed']== True)]
```

```python
#conversion are the people who completed the trial and their payment period is more than or equall to 1
#conversion rate in group_no_trial_fee
conversion_no_trial_fee=Group_no_trial_fee[(Group_no_trial_fee['payment_period'] >= 1)].subid.count()/Group_no_trial_fee.subid.count()
print(conversion_no_trial_fee)
```

```python
#conversion rate in group_low_trial_fee
conversion_low_trial_fee=Group_low_trial_fee[(Group_low_trial_fee['payment_period'] >= 1)].subid.count()/Group_low_trial_fee.subid.count()
print(conversion_low_trial_fee)
```

```python
#null hypo: p_middle <= p_low
#alternative hypo: p_middle>p_low
import math
import scipy
import scipy.stats as st
```

```python
#z-score
n =Group_low_trial_fee.shape[0]

numerator = conversion_low_trial_fee-conversion_no_trial_fee
denominator = math.sqrt((conversion_no_trial_fee*(1-conversion_no_trial_fee))/n)

numerator / denominator

```

```python
# calculate z critical value = 1.6448536269514722
scipy.stats.norm.ppf(1-.05)
```

```python
#z-score > z critical value, reject null hypothesis.
#There is enough evidence (alpha = 5%) to support the claim that low  trial fee=0.0367 improves conversion rates.
```

```python
#z-score > z critical value, reject null hypothesis.
#There is enough evidence (alpha = 5%) to support the claim that middle trial fee=0.1101 improves conversion rates.
```
